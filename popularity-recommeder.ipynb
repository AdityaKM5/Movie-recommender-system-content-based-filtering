```python
newdf['tmdbId']=pd.to_numeric(newdf['tmdbId'],errors='coerce')
```

```python
link = link.merge(newdf,on ='tmdbId')
```

```python
link.head()
```

```python
num_rating_df= link.groupby('original_title').count()['rating'].reset_index()
num_rating_df
```

```python
avg_rating_df= link.groupby('original_title').mean()['rating'].reset_index()
avg_rating_df
```

```python
avg_rating_df.rename(columns={'rating':'avg_rating'},inplace=True)
```

```python
num_rating_df.rename(columns={'rating':'number_of_rating'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='original_title')
popular_df
```

```python
popular_df = popular_df[popular_df['number_of_rating']>=1500].sort_values('avg_rating',ascending=False).head(30)
```

```python
popular_df
```

```python
popular_df.merge(movie,on='original_title').drop_duplicates('original_title')
```

```python
popular_df = popular_df.merge(newdf,on='original_title').drop_duplicates('original_title')
```
